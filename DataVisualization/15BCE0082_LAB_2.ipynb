```python
#In this lab, you will use Python and related packages (Pandas, Matplotlib, etc) to analyze the
#provided MLB baseball datasets, compute relevant statistics, and create visualizations. Your
#analysis and visualization results will provide evidence for you to answer the following
#questions.

#15BCE0082 Voleti Ravi L53-L54 ASSIGNMENT II

from __future__ import print_function

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#TASK 1
# 1.1) Compute the total number of wins for each of the teams in MLB over 2010-2015, display
# them in a descending sorted order.

df = pd.read_csv("teams.csv")
df = df[['yearID','teamID','W','L']]
df = df[(df['yearID'] >= 2010) & (df['yearID'] <= 2015)]
df = df.groupby('teamID', as_index=False)['W','L'].sum()
df['W/L'] = df['W']/df['L']
df1 = df.sort_values(by=['W/L'],ascending = False)
df1
```

```python
#TASK 1
# 1.2) Compute the average payroll per year for all teams over 2010-2015, displaying them in a
# descending sorted order

df = pd.read_csv("players.csv")
df = df[['teamID','yearID','salary']]
df = df[(df['yearID'] >= 2010) & (df['yearID'] <= 2015)]
df = df.groupby('teamID', as_index=False)['salary'].mean()
df2 = df.sort_values(by=['salary'],ascending = False)
df2
```

```python
# 1.3) Create a visualization of your choice which will all allow you to show whether a teams
# winning record is related to its payroll.

task1df = pd.merge(df1,df2, how='outer')
task1df = task1df[['W/L','salary']]
task1df = task1df.sort_values(by=['salary'],ascending = True)
graph = task1df.plot(x='salary',y='W/L',use_index='teamID',kind = 'line',title = "Average Payroll Vs Wins",figsize=(15,10))
plt.show()

# plotting the values in the form of bar graph proves that with an
# increas in amount of payroll does increase the winning chance of a 
# team.
```

```python
# Task 2
# 2.1) Compute the Batting Averages for each of the MLB teams over 2010-2015, display them
# in a descending sorted order. The Batting Average is defined as Hits/At Bats. The average is
# calculated from all players in each team.

df = pd.read_csv("players.csv")
df = df[['H','teamID','AB','yearID']]
df = df[(df['yearID'] >= 2010) & (df['yearID'] <= 2015)]
df = df.groupby('teamID', as_index=False)['H','AB'].sum()
df['ratio'] = df['H']/df['AB']
df3 = df.sort_values(by=['ratio'],ascending = False)
df3
```

```python
# 2.2) Create a visualization of your choice which will allow you to decide whether a teams batting
# average is related to its win-loss record.
df3 = df3[['ratio','teamID']]
task2df = pd.merge(df1,df3, how='outer')
task2df = task2df[['W/L','ratio']]
task2df = task2df.sort_values(by=['ratio'],ascending = True)
graph = task2df.plot(x='ratio',y='W/L',use_index='teamID',kind = 'line',title = "Batting Ratio Vs Wins",figsize=(15,10))
plt.show()

# We see that even with the increase in order of batting ratio there is an
# upwards trend in winning a match .
```

```python
# Task 3
# 3.1) Display the average ERA (Earned Run Average) for each of the MLB teams in 2010-2015,
# in a descending sorted order. A lower ERA indicates a better pitching performance.

df = pd.read_csv("pitching.csv")
df = df[['teamID','yearID','ERA']]
df = df[(df['yearID'] >= 2010) & (df['yearID'] <= 2015)]
df = df.groupby('teamID', as_index=False)['ERA'].mean()
df4 = df.sort_values(by=['ERA'],ascending = False)
df4
```

```python
# 3.2) Create a visualization of your choice which will allow you to decide if a teams win-loss
# record is related to its pitching performance.

task3df = pd.merge(df1,df4, how='outer')
task3df = task3df[['W/L','ERA']]
task3df = task3df.sort_values(by=['ERA'],ascending = True)
graph = task3df.plot(x='ERA',y='W/L',use_index='teamID',kind = 'bar',title = "ERA Vs Wins",figsize=(15,10))
plt.show()

# lower level of ERA means better pitching performance, as we see that
# for lower levels there is a moderdately high win rate which is gradually
# decreasing until we reach an ERA of 5.5 therfore there is some relation 
# between the pitching performance and win rate.
```

```python
# Task 4
# Using visualization to show the following trend: For the top 5 teams that have the most total
# wins between 2010 and 2015:

# 4.1) How are their Batting Averages changed from 2010-2015

df = pd.read_csv("players.csv")
df = df[['H','teamID','AB','yearID']]
newdf0 = df[df['yearID'] == 2010]
newdf1 = df[df['yearID'] == 2011]
newdf2 = df[df['yearID'] == 2012]
newdf3 = df[df['yearID'] == 2013]
newdf4 = df[df['yearID'] == 2014]
newdf5 = df[df['yearID'] == 2015]

newdf0 = newdf0.groupby('teamID', as_index=False)['H','AB'].sum()
newdf0['ratio 2010'] = newdf0['H']/newdf0['AB']
newdf1 = newdf1.groupby('teamID', as_index=False)['H','AB'].sum()
newdf1['ratio 2011'] = newdf1['H']/newdf1['AB']
newdf2 = newdf2.groupby('teamID', as_index=False)['H','AB'].sum()
newdf2['ratio 2012'] = newdf2['H']/newdf2['AB']
newdf3 = newdf3.groupby('teamID', as_index=False)['H','AB'].sum()
newdf3['ratio 2013'] = newdf3['H']/newdf3['AB']
newdf4 = newdf4.groupby('teamID', as_index=False)['H','AB'].sum()
newdf4['ratio 2014'] = newdf4['H']/newdf4['AB']
newdf5 = newdf5.groupby('teamID', as_index=False)['H','AB'].sum()
newdf5['ratio 2015'] = newdf5['H']/newdf5['AB']

task4df = pd.merge(newdf0,newdf1,on='teamID',how='outer')
task4df = pd.merge(task4df,newdf2,on='teamID',how='outer')
task4df = pd.merge(task4df,newdf3,on='teamID',how='outer')
task4df = pd.merge(task4df,newdf4,on='teamID',how='outer')
task4df = pd.merge(task4df,newdf5,on='teamID',how='outer')
task4df = task4df[['teamID','ratio 2010','ratio 2011','ratio 2012','ratio 2013','ratio 2014','ratio 2015']]
graph = task4df.plot(x='teamID',kind = 'bar',stacked = True,figsize=(30,10))
plt.show()
```
