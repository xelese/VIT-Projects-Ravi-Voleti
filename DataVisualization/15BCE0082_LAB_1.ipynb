```python
#12) List the average revenue, profit, and employee size for each
# industry.
arr = df['Industry'].unique()
length = len(arr)
profit = 0
revenue = 0
employee = 0
for i in range(0,length):
    #print(df.ix[df['Industry'] == arr[i],'Profits'])
    Pro = df.ix[df['Industry'] == arr[i],'Profits'].tolist()
    Avg = df.ix[df['Industry'] == arr[i],'Revenue'].tolist()
    Emp = df.ix[df['Industry'] == arr[i],'Employees'].tolist()
    for j in range(0,len(Pro)):
        profit += Pro[j]
    for j in range(0,len(Avg)):
        revenue += Avg[j]
    for j in range(0,len(Emp)):
        employee += Emp[j]
    #print(Pro)
    print(arr[i])
    print("Average Revenue: ",revenue/len(Avg),"Profit: ",profit,"Employees: ",employee)
    print("\n")
```

```python
#13) Group the companies by sector and industry, list their average
# revenue, profit, and employee size.
import pandas as pd
df = pd.read_csv("fortune1000.csv")
profit = 0
revenue = 0
employee = 0
for Company in df.groupby(['Sector','Industry']): #for Company in df.groupby('Industry'):
    Pro = df['Profits'].tolist()
    Avg = df['Revenue'].tolist()
    Emp = df['Employees'].tolist()
    for j in range(0,len(Pro)):
        profit += Pro[j]
    for j in range(0,len(Avg)):
        revenue += Avg[j]
    for j in range(0,len(Emp)):
        employee += Emp[j]
    print(Company)
    print("\n")
    print("Average Revenue: ",revenue/len(Avg),"Profit: ",profit,"Employees: ",employee)
    print("\n")
```

```python
#14) Find the top 10 industries based on the average profits/revenue ratio?
from __future__ import print_function

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

newdf = pd.read_csv("fortune1000.csv")
newdf = newdf[['Industry','Profits','Revenue']]
newdf
arr = newdf['Industry'].unique()
Cmparr = {}
Ratarr = {}
length = len(arr)
Profit = 0.00
Revenue = 0.00
Ratio = 0.00
for i in range(0,length):
    Pro = newdf.ix[newdf['Industry'] == arr[i],'Profits'].tolist()
    Rev = newdf.ix[newdf['Industry'] == arr[i],'Revenue'].tolist()
    for j in range(0,len(Pro)):
        Profit += Pro[j]
    for j in range(0,len(Rev)):
        Revenue += Rev[j]
    Ratio = Profit/Revenue
    #print(Ratio)
    Cmparr[i] = arr[i]
    Ratarr[i] = Ratio

adf = pd.DataFrame({'Industry':Cmparr, 'Ratio':Ratarr})
adf = adf.nlargest(10,'Ratio')
adf
```
